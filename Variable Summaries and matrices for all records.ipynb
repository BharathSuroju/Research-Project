```python
tf1.columns = ['words','tf']
```

```python
import numpy as np

for i,word in enumerate(tf1['words']):
  tf1.loc[i, 'idf'] = np.log(df_veris.shape[0]/(len(df_veris[df_veris['summary1'].str.contains(word)])))

tf1
```

```python
tf1['tfidf'] = tf1['tf'] * tf1['idf']
tf1
```

```python
tf1.sort_values(by='tfidf',ascending = False)[:50]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

crps = []
for review in df_veris['summary1']:
  crps.append(review)

vector = TfidfVectorizer(input=crps,min_df= 5,max_df= 0.5, stop_words='english')
mtx = vector.fit_transform(crps)
ftre = vector.get_feature_names()
idx=[n for n in range(1, len(crps)+1)]

tf_df=pd.DataFrame(mtx.T.todense(), index=ftre, columns=idx)
tf_df.to_csv('tfid.csv')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
word_vectorizer = CountVectorizer(ngram_range=(2,2), analyzer='word')
sparse_matrix = word_vectorizer.fit_transform(df_veris['summary1'])
frequencies = sum(sparse_matrix).toarray()[0]
freq = pd.DataFrame(frequencies, index=word_vectorizer.get_feature_names(), columns=['frequency'])
freq.sort_values(by='frequency',ascending = False)[:50]
```
